```python
import mysql.connector
from mysql.connector import connect, Error

config = {
'user': 'root',
'password': 'admin123',
'host': 'localhost',
'port': '3306',
'database': 'tradebot',
'raise_on_warnings': True,}


cnx = mysql.connector.connect(**config)    
cursor = cnx.cursor()

table = "tb_trades"
prefix = "TRD"
base_col = "TRD_CODE"
no_prec = 10

try:

    query = """
       SELECT %s FROM  %s ORDER BY RIGHT (%s,%s) DESC LIMIT 1;
    """% (
            base_col,
            table,
            base_col,
            no_prec,
        )
    

    cursor.execute(query)

    results = cursor.fetchone()
    
    for x in results:
        result = x
    
    result_str = result.strip('TRD')
    
    rawcount = int(result_str)
    rawcount = rawcount + 1
    multiplier = no_prec - len(str(rawcount))
    multiplier = 0 if multiplier <=0 else multiplier 
    code = "0"*multiplier + str(rawcount)
    print(prefix+code)
#     print({
#         "data" : result,
#         "msg" : "Successful Logged Activity",
#     })    

except Error as e:

    print({
        "data" : "Error",
        "msg" : e
    })  
       
```
